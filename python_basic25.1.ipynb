```python
def equal(l):
    if len(l)==3:
        c = []
        for i in l:
            d = l.count(i)
            c.append(d)
        c=sorted(c)
        maxn = c[2]
        
        if maxn ==1:
            return 0
        else:
            return maxn
    else:
        return "function takes three arguments only"
    
a = [3,4,3]
print(equal(a))
```
